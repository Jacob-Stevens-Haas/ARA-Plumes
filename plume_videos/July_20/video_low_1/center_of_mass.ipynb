# Apply Ensemble SINDy

## Frames [159, 260]

```python
import pysindy as ps 
from sklearn.preprocessing import StandardScaler
```

```python
subset_coeff = poly_coeff[159:260,:]
subset_coeff
```

```python
fps = 123
frame_count = len(subset_coeff)
start_time = 0
end_time = frame_count/fps
t = np.linspace(start_time,end_time,frame_count)
```

```python
# Instantiate and fit the SINDy model
feature_names = ['a', 'b', 'c']
ensemble_optimizer = ps.STLSQ()
model = ps.SINDy(feature_names=feature_names, optimizer=ensemble_optimizer)
# Ensemble with replacement (V1)
model.fit(subset_coeff, t=t, ensemble=True, quiet=True)
model.print()
ensemble_coefs = model.coef_list
```
